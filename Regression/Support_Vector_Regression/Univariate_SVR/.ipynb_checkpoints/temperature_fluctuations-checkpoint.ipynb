# Daily Temperature Fluctuations for La Crosse, Wisconsin

###                   Temperature measured daily over 1,095 days

## Import The Libraries 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVR
```

## Import Data Set

```python
#Data is imported and insepcted for null values 
heart = pd.read_csv("temperature_variation_dataset.csv")
heart.info()
```

## Split the Data into x and y Variables 

```python
x = heart.iloc[:,:-1].values
y = heart.iloc[:,-1].values

#The dependent variable data is reshaped to a 2D array for feature scaling
y = y.reshape(-1,1)
```

## Split the Data into Training and Test Subsets 

```python
#Data is split into a training set, which will be used to train the mode, and a testing set, used to test the model's accuracy 
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 0)
```

## Feature Scale the x and y Data Using Standardization

```python
#Two StandardScaler objects are instantiated, one for x data and one for y data 
sc_x = StandardScaler()
sc_y = StandardScaler()

#Standardization calculations are performed on x_train and y_train respectively, using their dedicated StandardScaler objects 
#The .fit_transform() method both performs these calculations and actualizes the transformation of the data
#The test portions of the data are then transformed using the same calculations, thereby ensuring we retain a consistant mathematical 
#relationship across training and test subsets. For this reason, we do not refit the StandardScaler object when scaling test-data. 

x_train = sc_x.fit_transform(x_train)
x_test = sc_x.transform(x_test)

y_train = sc_y.fit_transform(y_train)
y_test = sc_y.transform(y_test)
```

## Instantiate and Train a Support Vector Regression Model 

```python
#An SVR model is instantiated, assigning the Gaussian Radial Basis Function as its kernel
svr_regressor = SVR(kernel = 'rbf')

#y_train = y_train.ravel()
svr_regressor.fit(x_train,y_train)
svr_regressor
```

## Visualize Model Predictions vs. Training Data

```python
# An evenly spaced array of values is generated spanning the data's range of x-values
# This allows for smoother modeling of corresponding predictions
# the array is converted to a 2D array/grid
x_grid = np.arange(min(sc_x.inverse_transform(x_train)), 
                   max(sc_x.inverse_transform(x_train)), 
                   0.1).reshape(-1, 1)

#training data is converted to its orginal scale for proper visualization 
reverse_scaled_xtrain = sc_x.inverse_transform(x_train)
reverse_scaled_ytrain = sc_y.inverse_transform(y_train)

# Predictions are generated by the model based on points in the grid
y_pred_scaled = svr_regressor.predict(sc_x.transform(x_grid))
y_pred = sc_y.inverse_transform(y_pred_scaled.reshape(-1, 1)).ravel()

#Boundaries for the epsilon insensitive tube are defined here, the distance of its edges from the modeled predictions are 
#being equal to +/- the epsilon parameter of the SVR model
epsilon_high = sc_y.inverse_transform((y_pred_scaled + svr_regressor.epsilon).reshape(-1, 1)).ravel()
epsilon_low = sc_y.inverse_transform((y_pred_scaled - svr_regressor.epsilon).reshape(-1, 1)).ravel()


plt.figure(figsize=(10, 6))
plt.scatter(reverse_scaled_xtrain, reverse_scaled_ytrain, color='red', label='Real Data')
plt.plot(x_grid, y_pred, color='blue', label='Predicted Values')
plt.fill_between(x_grid.ravel(), epsilon_low, epsilon_high, color='yellow', alpha=0.6, label='Epsilon Insensitive Tube')
plt.grid(True, linestyle = '--', color = 'gray', linewidth = 0.5)
plt.xlabel('Day')
plt.ylabel('Temperature (C)')
plt.title('Temperature Fluctuations Over Time (Training Data)')
plt.legend(loc = 'lower left', fontsize = 'small')
plt.show()
```

## Visualize Model Predictions vs. Test Data

```python
x_grid = np.arange(min(sc_x.inverse_transform(x_train)), 
                   max(sc_x.inverse_transform(x_train)), 
                   0.1).reshape(-1, 1)

reverse_scaled_xtest = sc_x.inverse_transform(x_test)
reverse_scaled_ytest = sc_y.inverse_transform(y_test)


y_pred_scaled = svr_regressor.predict(sc_x.transform(x_grid))
y_pred = sc_y.inverse_transform(y_pred_scaled.reshape(-1, 1)).ravel()


epsilon_high = sc_y.inverse_transform((y_pred_scaled + svr_regressor.epsilon).reshape(-1, 1)).ravel()
epsilon_low = sc_y.inverse_transform((y_pred_scaled - svr_regressor.epsilon).reshape(-1, 1)).ravel()


plt.figure(figsize=(10, 6))
plt.scatter(reverse_scaled_xtest, reverse_scaled_ytest, color='red', label='Real Data')
plt.plot(x_grid, y_pred, color='blue', label='Predicted Values')
plt.fill_between(x_grid.ravel(), epsilon_low, epsilon_high, color='yellow', alpha=0.6, label='Epsilon Insensitive Tube')
plt.grid(True, linestyle = '--', color = 'gray', linewidth = 0.5)
plt.xlabel('Day')
plt.ylabel('Temperature (C)')
plt.title('Temperature Fluctuations Over Time (Test Data)')
plt.legend(loc = 'lower left', fontsize = 'small')
plt.show()
```

## Retrain the Model with Parameter Adjustments

```python
#Visualization demonstrates the model's inaccurate predictions at inflection points as temperature trends begin to reverse 
#The model will be adjusted in terms of Regularization (C), Epsilon, and Kernel Coefficient (gamma) and retrained

#Regularization is increased as little as possible to avoid overfitting the data
#Epsilon is increased slightly to account for noise
#The Kernel Coefficient is reset to increase the significance of each data point while still allowing the model to be generalizable

svr_regressor = SVR(kernel = 'rbf',  C = 20, epsilon = 0.18, gamma = 1.35)
svr_regressor.fit(x_train,y_train)
```

## Visualize the Retrained Model

```python
x_grid = np.arange(min(sc_x.inverse_transform(x_train)), 
                   max(sc_x.inverse_transform(x_train)), 
                   0.1).reshape(-1, 1)

reverse_scaled_xtrain = sc_x.inverse_transform(x_train)
reverse_scaled_ytrain = sc_y.inverse_transform(y_train)

y_pred_scaled = svr_regressor.predict(sc_x.transform(x_grid))
y_pred = sc_y.inverse_transform(y_pred_scaled.reshape(-1, 1)).ravel()


epsilon_high = sc_y.inverse_transform((y_pred_scaled + svr_regressor.epsilon).reshape(-1, 1)).ravel()
epsilon_low = sc_y.inverse_transform((y_pred_scaled - svr_regressor.epsilon).reshape(-1, 1)).ravel()


plt.figure(figsize=(10, 6))
plt.scatter(reverse_scaled_xtrain, reverse_scaled_ytrain, color='red', label='Real Data')
plt.plot(x_grid, y_pred, color='blue', label='Predicted Values')
plt.fill_between(x_grid.ravel(), epsilon_low, epsilon_high, color='yellow', alpha=0.6, label='Epsilon Insensitive Tube')
plt.grid(True, linestyle = '--', color = 'gray', linewidth = 0.5)
plt.xlabel('Day')
plt.ylabel('Temperature (C)')
plt.title('Temperature Fluctuations Over Time (Training Data)')
plt.legend(loc = 'lower left', fontsize = 'small')
plt.show()
```

```python
x_grid = np.arange(min(sc_x.inverse_transform(x_train)), 
                   max(sc_x.inverse_transform(x_train)), 
                   0.1).reshape(-1, 1)

reverse_scaled_xtest = sc_x.inverse_transform(x_test)
reverse_scaled_ytest = sc_y.inverse_transform(y_test)


y_pred_scaled_test = svr_regressor.predict(sc_x.transform(x_grid))
y_pred_test = sc_y.inverse_transform(y_pred_scaled_test.reshape(-1, 1)).ravel()


epsilon_high = sc_y.inverse_transform((y_pred_scaled_test + svr_regressor.epsilon).reshape(-1, 1)).ravel()
epsilon_low = sc_y.inverse_transform((y_pred_scaled_test - svr_regressor.epsilon).reshape(-1, 1)).ravel()


plt.figure(figsize=(10, 6))
plt.scatter(reverse_scaled_xtest, reverse_scaled_ytest, color='red', label='Real Data')
plt.plot(x_grid, y_pred_test, color='blue', label='Predicted Values')
plt.fill_between(x_grid.ravel(), epsilon_low, epsilon_high, color='yellow', alpha=0.6, label='Epsilon Insensitive Tube')
plt.grid(True, linestyle = '--', color = 'gray', linewidth = 0.5)
plt.xlabel('Day')
plt.ylabel('Temperature (C)')
plt.title('Temperature Fluctuations Over Time (Test Data)')
plt.legend(loc = 'lower left', fontsize = 'small')
plt.show()
```

## Calculate and Visualize Residuals between Model and Test Data

```python
#Predicted temperature values are subtracted from real values to calculate residual
#Residuals are then plotted against the independent time variable, 
#Residuals appear to be randomly distibuted around the zero residual reference line 

y_actual = sc_y.inverse_transform(y_test)
y_predictions = sc_y.inverse_transform(svr_regressor.predict(x_test).reshape(-1,1))
residuals = y_actual - y_predictions

plt.figure()
plt.scatter(sc_x.inverse_transform(x_test),residuals, color = "blue")
plt.xlabel("Day")
plt.ylabel("Residuals")
plt.title("Residuals for Test Data vs. Predictions")
plt.grid(True, linestyle = '--', color = 'gray', linewidth = 0.5)
plt.axhline(y = 0, linestyle = "--", linewidth= 1.5, color = 'red', label = "Zero Residual Line")
plt.legend()
plt.show()



```

## Evaluate the Percent Accuracy of Predictions

```python
#predicted values (derived from x_test data as input) are concatenated together with actual y_test values, creating a 2D array
#The resulting 2D array contains a prediction and an actual value for each inner array

np.set_printoptions(precision = 2)
predictions_vs_testData = np.concatenate((y_actual,y_predictions), axis = 1)
#print(predictions_vs_testData)

#The code below iterates through the concatenated array and calculates the percent closeness of the two numbers 
#A new series is generated representing the percent accuracy of each prediction in comparison to the actual data value 
#The .describe() method is used to generate basic descriptive statistics about the predictice accuracy of the model 
percent_accuracy = []
for i in predictions_vs_testData:
    smaller = min(abs(i[0]),abs(i[1]))
    larger = max(abs(i[0]),abs(i[1]))
    percent_accuracy.append((smaller/larger)*100)

#Calling describe on percent_accuracy shows a mean predicted accuracy of approximately 81.79 percent and a median of 90.23 percent
#This can however be deceptive considering the data's level of noise 
#ratios between two very small numbers can also appear more signficant than is actually the case
#For example, if a real value was 2 and the model predicted 1. This would represent only 50% accuracy in spite of being very close
#to the real value 
test_percent_accuracy = pd.Series(percent_accuracy)
test_percent_accuracy.describe()
```

## Calculate the Mean Absolute Error of the Model's Predictions

```python
#When calculating the mean absolute error, we find that the average error between prediction and real data is actually very small
#There is a mean discrepancy of only 1.61 degrees Celsius between prediction and real values 

mean_abs_error = np.mean(np.abs(predictions_vs_testData[:, 0] - predictions_vs_testData[:, 1]))
print(f"\nMean Absolute Error (MAE): {mean_abs_error:.2f}")
```

## Calculate the Root Mean Square Error of the Model's Predictions

```python
#The Root Mean Square Error is comparable to the Mean Absolute Error
#The RMSE implies that predictions deviate from real values by an average of only 2.01 degrees Celsius 
rmse = round(np.sqrt(np.mean((predictions_vs_testData[:, 0] - predictions_vs_testData[:, 1]) ** 2)),2)
print(f"\nRoot Mean Squared Error (RMSE): {rmse:.2f}")
```

## Predict the Temperature on a Future Day

```python
prediction_value = sc_x.transform([[1600]])
result = sc_y.inverse_transform(svr_regressor.predict(prediction_value).reshape(-1,1))
result = round(result[0][0],2)

print(f"The temperature on this day is predicted to be {result} degrees Celsius with a potential error of +/- {rmse} degrees.")
```
